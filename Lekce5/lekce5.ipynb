````python
from collections import deque


class TreeNode(object):
    """NejjednoduÅ¡Å¡Ã­ uzel binÃ¡rnÃ­ho stromu â€“ stejnÃ© jako na LeetCode."""

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(values):
    """PosklÃ¡dÃ¡ strom z ÃºrovÅˆovÃ©ho seznamu hodnot (None = chybÄ›jÃ­cÃ­ uzel)."""
    if not values:
        return None

    root = TreeNode(values[0])
    queue = deque([root])
    index = 1

    while queue and index < len(values):
        node = queue.popleft()

        if index < len(values) and values[index] is not None:
            node.left = TreeNode(values[index])
            queue.append(node.left)
        index += 1

        if index < len(values) and values[index] is not None:
            node.right = TreeNode(values[index])
            queue.append(node.right)
        index += 1

    return root


def level_order(root):
    """VrÃ¡tÃ­ strom jako seznam hodnot po ÃºrovnÃ­ch (uÅ¾iteÄnÃ© pro kontrolu)."""
    if not root:
        return []

    result = []
    queue = deque([root])

    while queue:
        node = queue.popleft()
        if node is None:
            result.append(None)
            continue

        result.append(node.val)
        queue.append(node.left)
        queue.append(node.right)

    while result and result[-1] is None:
        result.pop()
    return result
````
