## 🧮 BONUS: Evaluate Reverse Polish Notation



**Obtížnost:** 🟡 Medium | **Časová složitost:** O(n) | **Prostorová složitost:** O(n)



### 📋 **Zadání:**

**Poznámky:**

- Platné operátory jsou `'+'`, `'-'`, `'*'`, a `'/'`

- Dělení mezi dvěma celými čísly vždy zakřůlí směrem k nule

- Nebude dělení nulou

- Vstup reprezentuje platný RPN výraz



### 💡 **Co je Reverse Polish Notation?**

RPN (také postfix notace) umítì operátory **za** operandy:

- Normální: `(2 + 1) * 3`

- RPN: `2 1 + 3 *`
