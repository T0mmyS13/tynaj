## 3Ô∏è‚É£ ≈òe≈°en√≠ √∫kolu 4.1 ‚Äì Valid Palindrome



**Obt√≠≈ænost:** üü¢ Easy | **ƒåasov√° slo≈æitost:** O(n) | **Prostorov√° slo≈æitost:** O(1)



### üí≠ My≈°lenkov√Ω proces

1. **Naivn√≠ p≈ô√≠stup:** Vytvo≈ôit vyfiltrovan√Ω a normalizovan√Ω ≈ôetƒõzec, pak porovnat s jeho reverz√≠ ‚Üí O(n) ƒças, O(n) prostor.

2. **Two Pointers:** Porovn√°vat znaky p≈ô√≠mo od kraj≈Ø, p≈ôeskoƒçit ne-alfanumerick√© znaky ‚Üí O(n) ƒças, O(1) prostor.

3. **Optimalizace:** Pou≈æ√≠t vestavbu `isalnum()` a `lower()` pro ƒçistou implementaci.



### üéØ Kl√≠ƒçov√© poznatky

- Two Pointers umo≈æ≈àuj√≠ in-place porovn√°v√°n√≠ bez extra pamƒõti.

- Okrajov√© p≈ô√≠pady: pr√°zdn√Ω ≈ôetƒõzec, pouze interpunkce, unicode znaky.

- P≈ôeskoƒçen√≠ ne-alfanumerick√Ωch znak≈Ø je kl√≠ƒçov√© pro spr√°vnou funkci.

```python
class ValidPalindromeSolution(object):

    """Hlavn√≠ ≈ôe≈°en√≠ ‚Äì Two Pointers in-place."""



    def isPalindrome(self, s: str) -> bool:

        left, right = 0, len(s) - 1

        

        while left < right:

            # P≈ôeskoƒç ne-alfanumerick√© znaky zleva

            while left < right and not s[left].isalnum():

                left += 1

            

            # P≈ôeskoƒç ne-alfanumerick√© znaky zprava

            while left < right and not s[right].isalnum():

                right -= 1

            

            # Porovnej znaky (case-insensitive)

            if s[left].lower() != s[right].lower():

                return False

            

            left += 1

            right -= 1

        

        return True





class ValidPalindromeNaiveSolution(object):

    """Alternativn√≠ ≈ôe≈°en√≠ ‚Äì vytvo≈ôen√≠ nov√©ho ≈ôetƒõzce."""



    def isPalindrome(self, s: str) -> bool:

        # Normalizace: pouze alfanumerick√© znaky, mal√° p√≠smena

        cleaned = ''.join(char.lower() for char in s if char.isalnum())

        return cleaned == cleaned[::-1]





class ValidPalindromeRegexSolution(object):

    """Varianta s regul√°rn√≠mi v√Ωrazy."""



    def isPalindrome(self, s: str) -> bool:

        import re

        # Odebr√°n√≠ v≈°ech ne-alfanumerick√Ωch znak≈Ø

        cleaned = re.sub(r'[^a-zA-Z0-9]', '', s.lower())

        return cleaned == cleaned[::-1]





main_valid_palindrome = ValidPalindromeSolution()

naive_valid_palindrome = ValidPalindromeNaiveSolution()

regex_valid_palindrome = ValidPalindromeRegexSolution()



print("‚úÖ ≈òe≈°en√≠ pro Valid Palindrome p≈ôipravena")
```

## 4Ô∏è‚É£ Testy pro √∫kol 4.1



Testujeme hlavn√≠ i alternativn√≠ implementace pomoc√≠ `unittest`. Sc√©n√°≈ôe pokr√Ωvaj√≠ r≈Øzn√© typy palindrom≈Ø, pr√°zdn√© ≈ôetƒõzce i slo≈æit√© kombinace interpunkce a unicode znak≈Ø.

```python
class TestValidPalindrome(unittest.TestCase):

    def setUp(self) -> None:

        self.solutions = [

            ("two_pointers", main_valid_palindrome.isPalindrome),

            ("naive_clean", naive_valid_palindrome.isPalindrome),

            ("regex_clean", regex_valid_palindrome.isPalindrome),

        ]



    def test_valid_palindrome_cases(self):

        for label, fn in self.solutions:

            with self.subTest(solution=label):

                for s, expected in valid_palindrome_cases:

                    result = fn(s)

                    self.assertEqual(

                        result,

                        expected,

                        msg=f"Solution '{label}' selhala pro vstup '{s}'",

                    )





suite_palindrome = unittest.TestLoader().loadTestsFromTestCase(TestValidPalindrome)

result_palindrome = unittest.TextTestRunner(verbosity=1).run(suite_palindrome)

assert result_palindrome.wasSuccessful(), "Testy pro Valid Palindrome nepro≈°ly!"
```

## 5Ô∏è‚É£ ≈òe≈°en√≠ √∫kolu 4.2 ‚Äì Valid Parentheses



**Obt√≠≈ænost:** üü¢ Easy | **ƒåasov√° slo≈æitost:** O(n) | **Prostorov√° slo≈æitost:** O(n)



### üí≠ My≈°lenkov√Ω proces

1. Stack (z√°sobn√≠k) je ide√°ln√≠ datovou strukturou pro sledov√°n√≠ vno≈ôen√Ωch struktur.

2. P≈ôi otev√≠rac√≠ z√°vorce p≈ôid√°me ji na stack.

3. P≈ôi uzav√≠rac√≠ z√°vorce zkontrolujeme, zda odpov√≠d√° posledn√≠ otev√≠rac√≠ z√°vorce na stacku.

4. Po projit√≠ cel√©ho ≈ôetƒõzce mus√≠ b√Ωt stack pr√°zdn√Ω.



### üéØ Kl√≠ƒçov√© poznatky

- Stack (LIFO - Last In, First Out) p≈ôesnƒõ modeluje vno≈ôen√© struktury.

- Slovn√≠k pro mapov√°n√≠ uzav√≠rac√≠ch na otev√≠rac√≠ z√°vorky zjednodu≈°uje logiku.

- Early termination p≈ôi prvn√≠ chybƒõ zvy≈°uje efektivitu.

```python
class ValidParenthesesSolution(object):

    """Hlavn√≠ ≈ôe≈°en√≠ ‚Äì Stack s mapov√°n√≠m."""



    def isValid(self, s: str) -> bool:

        # Mapov√°n√≠ uzav√≠rac√≠ch z√°vorek na odpov√≠daj√≠c√≠ otev√≠rac√≠

        mapping = {')': '(', '}': '{', ']': '['}

        stack = []

        

        for char in s:

            if char in mapping:  # Uzav√≠rac√≠ z√°vorka

                # Zkontroluj, zda stack nen√≠ pr√°zdn√Ω a zda se shoduje vrchol

                if not stack or stack.pop() != mapping[char]:

                    return False

            else:  # Otev√≠rac√≠ z√°vorka

                stack.append(char)

        

        # V≈°echny z√°vorky mus√≠ b√Ωt uzav≈ôeny

        return len(stack) == 0





class ValidParenthesesAlternativeSolution(object):

    """Alternativn√≠ ≈ôe≈°en√≠ ‚Äì explicitn√≠ kontrola typ≈Ø z√°vorek."""



    def isValid(self, s: str) -> bool:

        stack = []

        opening = {'(', '[', '{'}

        pairs = {'(': ')', '[': ']', '{': '}'}

        

        for char in s:

            if char in opening:

                stack.append(char)

            else:  # Uzav√≠rac√≠ z√°vorka

                if not stack:

                    return False

                last_opening = stack.pop()

                if pairs[last_opening] != char:

                    return False

        

        return len(stack) == 0





class ValidParenthesesCounterSolution(object):

    """Varianta s poƒç√≠t√°n√≠m ‚Äì funguje pouze pro jeden typ z√°vorek."""



    def isValid(self, s: str) -> bool:

        # Tato varianta funguje spr√°vnƒõ pouze pro jeden typ z√°vorek!

        if len(set(s)) > 2:  # V√≠ce ne≈æ jeden typ z√°vorek

            return False

        

        balance = 0

        for char in s:

            if char in '([{':

                balance += 1

            elif char in ')]}':

                balance -= 1

                if balance < 0:  # V√≠ce uzav√≠rac√≠ch ne≈æ otev√≠rac√≠ch

                    return False

        

        return balance == 0





main_valid_parentheses = ValidParenthesesSolution()

alt_valid_parentheses = ValidParenthesesAlternativeSolution()

counter_valid_parentheses = ValidParenthesesCounterSolution()



print("‚úÖ ≈òe≈°en√≠ pro Valid Parentheses p≈ôipravena")
```

## 6Ô∏è‚É£ Testy pro √∫kol 4.2



Testujeme v≈°echny t≈ôi varianty ≈ôe≈°en√≠. Pozor na to, ≈æe counter varianta nefunguje spr√°vnƒõ pro sm√≠≈°en√© typy z√°vorek ‚Äì toto je √∫mysln√© uk√°z√°n√≠ limitace jednodu≈°≈°√≠ho p≈ô√≠stupu.

```python
class TestValidParentheses(unittest.TestCase):

    def setUp(self) -> None:

        # Counter varianta nefunguje spr√°vnƒõ pro sm√≠≈°en√© z√°vorky - testujeme jen hlavn√≠ varianty

        self.solutions = [

            ("stack_mapping", main_valid_parentheses.isValid),

            ("stack_explicit", alt_valid_parentheses.isValid),

        ]



    def test_valid_parentheses_cases(self):

        for label, fn in self.solutions:

            with self.subTest(solution=label):

                for s, expected in valid_parentheses_cases:

                    result = fn(s)

                    self.assertEqual(

                        result,

                        expected,

                        msg=f"Solution '{label}' selhala pro vstup '{s}'",

                    )



    def test_counter_solution_limitations(self):

        """Uk√°≈æem, ≈æe counter ≈ôe≈°en√≠ nefunguje pro sm√≠≈°en√© z√°vorky."""

        # Tento test by mƒõl selhat pro counter ≈ôe≈°en√≠

        problematic_cases = [("([)]", False), ("()[]{}", True)]

        

        for s, expected in problematic_cases:

            result = counter_valid_parentheses.isValid(s)

            # Counter ≈ôe≈°en√≠ vrac√≠ nespr√°vn√© v√Ωsledky pro nƒõkter√© p≈ô√≠pady

            print(f"Counter pro '{s}': {result} (oƒçek√°v√°no: {expected})")





suite_parentheses = unittest.TestLoader().loadTestsFromTestCase(TestValidParentheses)

result_parentheses = unittest.TextTestRunner(verbosity=1).run(suite_parentheses)

assert result_parentheses.wasSuccessful(), "Testy pro Valid Parentheses nepro≈°ly!"
```

## üß† BONUS ≈ôe≈°en√≠ ‚Äì Evaluate Reverse Polish Notation



**Obt√≠≈ænost:** üü° Medium | **ƒåasov√° slo≈æitost:** O(n) | **Prostorov√° slo≈æitost:** O(n)



### üí≠ My≈°lenkov√Ω proces

1. RPN (Reverse Polish Notation) je postfix notace - oper√°tory n√°sleduj√≠ za operandy.

2. Stack je ide√°ln√≠ pro vyhodnocov√°n√≠ RPN - operandy ukl√°d√°me, oper√°tory je konzumuj√≠.

3. P≈ôi oper√°toru vezmeme dva posledn√≠ operandy ze stacku, provedeme operaci a v√Ωsledek vr√°t√≠me na stack.

4. Fin√°ln√≠ v√Ωsledek je jedin√Ω prvek zb√Ωvaj√≠c√≠ na stacku.



### üéØ Kl√≠ƒçov√© poznatky

- Po≈ôad√≠ operand≈Ø je d≈Øle≈æit√©: `a b -` znamen√° `a - b`, ne `b - a`.

- Dƒõlen√≠ v Pythonu automaticky zaokrouhli k nule pro z√°porn√© v√Ωsledky.

- Funkcion√°ln√≠ p≈ô√≠stup s `operator` modulem ƒçin√≠ k√≥d ƒçitelnƒõj≈°√≠m.

```python
class EvaluateRPNSolution(object):

    """Hlavn√≠ ≈ôe≈°en√≠ ‚Äì Stack s explicitn√≠mi operacemi."""



    def evalRPN(self, tokens: List[str]) -> int:

        stack = []

        

        for token in tokens:

            if token in ["+", "-", "*", "/"]:

                # Vezmi dva operandy (pozor na po≈ôad√≠!)

                b = stack.pop()

                a = stack.pop()

                

                if token == "+":

                    result = a + b

                elif token == "-":

                    result = a - b

                elif token == "*":

                    result = a * b

                elif token == "/":

                    # Python 3 dƒõl√≠ s float v√Ωsledkem, pot≈ôebujeme integer division toward zero

                    result = int(a / b)  # Automaticky zaokrouhl√≠ k nule

                

                stack.append(result)

            else:

                # Token je ƒç√≠slo

                stack.append(int(token))

        

        return stack[0]





class EvaluateRPNOperatorSolution(object):

    """Alternativn√≠ ≈ôe≈°en√≠ ‚Äì pou≈æit√≠ operator modulu."""



    def evalRPN(self, tokens: List[str]) -> int:

        stack = []

        operations = {

            "+": operator.add,

            "-": operator.sub,

            "*": operator.mul,

            "/": lambda a, b: int(a / b),  # Spec√≠n√≠ handling pro dƒõlen√≠

        }

        

        for token in tokens:

            if token in operations:

                b = stack.pop()

                a = stack.pop()

                stack.append(operations[token](a, b))

            else:

                stack.append(int(token))

        

        return stack[0]





class EvaluateRPNRecursiveSolution(object):

    """Rekurzivn√≠ ≈ôe≈°en√≠ pro srovn√°n√≠."""



    def evalRPN(self, tokens: List[str]) -> int:

        def helper(index):

            token = tokens[index]

            if token in ["+", "-", "*", "/"]:

                # Rekurzivnƒõ vyhodno≈• operandy

                right, next_index = helper(index - 1)

                left, next_index = helper(next_index - 1)

                

                if token == "+":

                    return left + right, next_index

                elif token == "-":

                    return left - right, next_index

                elif token == "*":

                    return left * right, next_index

                elif token == "/":

                    return int(left / right), next_index

            else:

                return int(token), index

        

        result, _ = helper(len(tokens) - 1)

        return result





main_evaluate_rpn = EvaluateRPNSolution()

operator_evaluate_rpn = EvaluateRPNOperatorSolution()

recursive_evaluate_rpn = EvaluateRPNRecursiveSolution()



print("‚úÖ ≈òe≈°en√≠ pro Evaluate RPN p≈ôipravena")
```
