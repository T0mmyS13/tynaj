## ğŸ” Å˜eÅ¡enÃ­ 1: Najdi ÄÃ­slo nejblÃ­Å¾ k nule

**ObtÃ­Å¾nost:** ğŸŸ¢ Easy | **ÄŒasovÃ¡ sloÅ¾itost:** O(n) | **ProstorovÃ¡ sloÅ¾itost:** O(1)

### ğŸ’­ **MyÅ¡lenkovÃ½ proces:**
1. PotÅ™ebujeme najÃ­t ÄÃ­slo s nejmenÅ¡Ã­ vzdÃ¡lenostÃ­ od nuly
2. VzdÃ¡lenost = `abs(ÄÃ­slo)`
3. PÅ™i stejnÃ© vzdÃ¡lenosti vybÃ­rÃ¡me vÄ›tÅ¡Ã­ ÄÃ­slo
4. Projdeme pole jednou a sledujeme nejlepÅ¡Ã­ kandidÃ¡t

### ğŸ¯ **KlÃ­ÄovÃ© pozorovÃ¡nÃ­:**
- StaÄÃ­ jeden prÅ¯chod polem â†’ O(n)
- PorovnÃ¡vÃ¡me vzdÃ¡lenosti, ale pÅ™i rovnosti bereme vÄ›tÅ¡Ã­ hodnotu
- HraniÄnÃ­ pÅ™Ã­pady: prÃ¡zdnÃ© pole (nebude), jeden prvek

```python
class Solution1(object):
    def findClosestNumber(self, nums):
        """
        HlavnÃ­ Å™eÅ¡enÃ­ - O(n) Äas, O(1) prostor
        """
        # Inicializujeme s prvnÃ­m prvkem
        closest = nums[0]
        
        # Projdeme vÅ¡echny prvky
        for num in nums:
            # Pokud aktuÃ¡lnÃ­ ÄÃ­slo je blÃ­Å¾e k nule
            if abs(num) < abs(closest):
                closest = num
            # Pokud majÃ­ stejnou vzdÃ¡lenost, vybereme vÄ›tÅ¡Ã­
            elif abs(num) == abs(closest) and num > closest:
                closest = num
                
        return closest

# ğŸ”„ AlternativnÃ­ Å™eÅ¡enÃ­ s vyuÅ¾itÃ­m min() funkce
class Solution1_Alternative(object):
    def findClosestNumber(self, nums):
        """
        AlternativnÃ­ Å™eÅ¡enÃ­ s min() funkcÃ­ - takÃ© O(n)
        """
        return min(nums, key=lambda x: (abs(x), -x))
        # Key funkce: (vzdÃ¡lenost_od_nuly, -hodnota)
        # -x zajiÅ¡Å¥uje, Å¾e pÅ™i stejnÃ© vzdÃ¡lenosti vybere vÄ›tÅ¡Ã­ ÄÃ­slo

# ğŸ§ª TestovÃ¡nÃ­ Å™eÅ¡enÃ­
def test_closest_number():
    print("ğŸ” TESTOVÃNÃ ÃšLOHY 1: Najdi ÄÃ­slo nejblÃ­Å¾ k nule")
    print("=" * 60)
    
    sol1 = Solution1()
    sol_alt = Solution1_Alternative()
    
    test_cases = [
        ([-4, -2, 1, 4, 8], 1),
        ([2, -1, 1], 1),
        ([-2, 2], 2),
        ([1], 1),
        ([-1], -1),
        ([0, 3, -2], 0)
    ]
    
    for i, (nums, expected) in enumerate(test_cases, 1):
        result1 = sol1.findClosestNumber(nums)
        result_alt = sol_alt.findClosestNumber(nums)
        
        print(f"Test {i}: nums = {nums}")
        print(f"  OÄekÃ¡vÃ¡no: {expected}")
        print(f"  HlavnÃ­ Å™eÅ¡enÃ­: {result1} {'âœ…' if result1 == expected else 'âŒ'}")
        print(f"  AlternativnÃ­: {result_alt} {'âœ…' if result_alt == expected else 'âŒ'}")
        print()
    
    print("ğŸ“Š AnalÃ½za sloÅ¾itosti:")
    print("  ÄŒasovÃ¡ sloÅ¾itost: O(n) - musÃ­me projÃ­t vÅ¡echny prvky")
    print("  ProstorovÃ¡ sloÅ¾itost: O(1) - pouÅ¾Ã­vÃ¡me konstantnÃ­ pamÄ›Å¥")
    print()

# SpuÅ¡tÄ›nÃ­ testÅ¯
test_closest_number()
```

## ğŸ›ï¸ Å˜eÅ¡enÃ­ 2: PÅ™evod Å™Ã­mskÃ½ch ÄÃ­slic na celÃ¡ ÄÃ­sla

**ObtÃ­Å¾nost:** ğŸŸ¢ Easy | **ÄŒasovÃ¡ sloÅ¾itost:** O(n) | **ProstorovÃ¡ sloÅ¾itost:** O(1)

### ğŸ’­ **MyÅ¡lenkovÃ½ proces:**
1. VytvoÅ™Ã­me mapovÃ¡nÃ­ symbolÅ¯ na hodnoty
2. Projdeme Å™etÄ›zec zleva doprava
3. Pokud aktuÃ¡lnÃ­ symbol mÃ¡ menÅ¡Ã­ hodnotu neÅ¾ nÃ¡sledujÃ­cÃ­ â†’ odeÄteme
4. Jinak pÅ™iÄteme hodnotu symbolu

### ğŸ¯ **KlÃ­ÄovÃ© pozorovÃ¡nÃ­:**
- VÄ›tÅ¡inou pÅ™iÄÃ­tÃ¡me, odeÄÃ­tÃ¡me jen v speciÃ¡lnÃ­ch pÅ™Ã­padech (IV, IX, XL, XC, CD, CM)
- StaÄÃ­ porovnat aktuÃ¡lnÃ­ a nÃ¡sledujÃ­cÃ­ symbol
- MÅ¯Å¾eme Å™eÅ¡it takÃ© zprava doleva (pak vÅ¾dy pÅ™iÄÃ­tÃ¡me nebo odeÄÃ­tÃ¡me)

```python
class Solution2(object):
    def romanToInt(self, s):
        """
        HlavnÃ­ Å™eÅ¡enÃ­ - prochÃ¡zenÃ­ zleva doprava, O(n) Äas
        """
        # MapovÃ¡nÃ­ Å™Ã­mskÃ½ch symbolÅ¯ na hodnoty
        roman_map = {
            'I': 1, 'V': 5, 'X': 10, 'L': 50,
            'C': 100, 'D': 500, 'M': 1000
        }
        
        total = 0
        i = 0
        
        while i < len(s):
            # Pokud nejsme na poslednÃ­m symbolu a aktuÃ¡lnÃ­ < nÃ¡sledujÃ­cÃ­
            if i + 1 < len(s) and roman_map[s[i]] < roman_map[s[i + 1]]:
                # OdeÄteme aktuÃ¡lnÃ­ hodnotu (subtraction case)
                total += roman_map[s[i + 1]] - roman_map[s[i]]
                i += 2  # PÅ™eskoÄÃ­me oba symboly
            else:
                # StandardnÃ­ pÅ™Ã­pad - pÅ™iÄteme hodnotu
                total += roman_map[s[i]]
                i += 1
                
        return total

# ğŸ”„ AlternativnÃ­ Å™eÅ¡enÃ­ - prochÃ¡zenÃ­ zprava doleva
class Solution2_Alternative(object):
    def romanToInt(self, s):
        """
        AlternativnÃ­ pÅ™Ã­stup - zprava doleva, jednoduÅ¡Å¡Ã­ logika
        """
        roman_map = {
            'I': 1, 'V': 5, 'X': 10, 'L': 50,
            'C': 100, 'D': 500, 'M': 1000
        }
        
        total = 0
        prev_value = 0
        
        # ProchÃ¡zÃ­me zprava doleva
        for char in reversed(s):
            value = roman_map[char]
            
            # Pokud aktuÃ¡lnÃ­ hodnota < pÅ™edchozÃ­ â†’ odeÄÃ­tÃ¡me
            if value < prev_value:
                total -= value
            else:
                total += value
                
            prev_value = value
            
        return total

# âš¡ OptimalizovanÃ© Å™eÅ¡enÃ­ s pÅ™edpoÄÃ­tanÃ½mi dvojicemi
class Solution2_Optimized(object):
    def romanToInt(self, s):
        """
        OptimalizovanÃ© Å™eÅ¡enÃ­ s pÅ™edpoÄÃ­tanÃ½mi subtraction pÅ™Ã­pady
        """
        # VÄetnÄ› speciÃ¡lnÃ­ch pÅ™Ã­padÅ¯
        roman_map = {
            'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000,
            'IV': 4, 'IX': 9, 'XL': 40, 'XC': 90, 'CD': 400, 'CM': 900
        }
        
        total = 0
        i = 0
        
        while i < len(s):
            # ZkusÃ­me nejdÅ™Ã­ve dvouznakovÃ½ symbol
            if i + 1 < len(s) and s[i:i+2] in roman_map:
                total += roman_map[s[i:i+2]]
                i += 2
            else:
                total += roman_map[s[i]]
                i += 1
                
        return total

# ğŸ§ª TestovÃ¡nÃ­ vÅ¡ech Å™eÅ¡enÃ­
def test_roman_to_int():
    print("ğŸ›ï¸ TESTOVÃNÃ ÃšLOHY 2: PÅ™evod Å™Ã­mskÃ½ch ÄÃ­slic")
    print("=" * 60)
    
    sol1 = Solution2()
    sol_alt = Solution2_Alternative() 
    sol_opt = Solution2_Optimized()
    
    test_cases = [
        ("III", 3),
        ("LVIII", 58),
        ("MCMXCIV", 1994),
        ("IV", 4),
        ("IX", 9),
        ("XL", 40),
        ("XC", 90),
        ("CD", 400),
        ("CM", 900),
        ("MCDXLIV", 1444)  # 1000 + 400 + 40 + 4
    ]
    
    for i, (roman, expected) in enumerate(test_cases, 1):
        result1 = sol1.romanToInt(roman)
        result_alt = sol_alt.romanToInt(roman)
        result_opt = sol_opt.romanToInt(roman)
        
        print(f"Test {i}: '{roman}' â†’ {expected}")
        print(f"  HlavnÃ­ Å™eÅ¡enÃ­: {result1} {'âœ…' if result1 == expected else 'âŒ'}")
        print(f"  AlternativnÃ­: {result_alt} {'âœ…' if result_alt == expected else 'âŒ'}")
        print(f"  OptimalizovanÃ©: {result_opt} {'âœ…' if result_opt == expected else 'âŒ'}")
        print()
    
    print("ğŸ“Š PorovnÃ¡nÃ­ pÅ™Ã­stupÅ¯:")
    print("  1. Zleva doprava: IntuitivnÃ­, ale sloÅ¾itÄ›jÅ¡Ã­ podmÃ­nky")
    print("  2. Zprava doleva: JednoduÅ¡Å¡Ã­ logika, stejnÃ¡ sloÅ¾itost")
    print("  3. PÅ™edpoÄÃ­tanÃ© dvojice: NejrychlejÅ¡Ã­ v praxi")
    print("  VÅ¡echny: O(n) Äas, O(1) prostor")
    print()

# SpuÅ¡tÄ›nÃ­ testÅ¯
test_roman_to_int()
```

## ğŸ”„ BONUS Å˜eÅ¡enÃ­: SlouÄenÃ­ pÅ™ekrÃ½vajÃ­cÃ­ch se intervalÅ¯

**ObtÃ­Å¾nost:** ğŸŸ¡ Medium | **ÄŒasovÃ¡ sloÅ¾itost:** O(n log n) | **ProstorovÃ¡ sloÅ¾itost:** O(1) aÅ¾ O(n)

### ğŸ’­ **MyÅ¡lenkovÃ½ proces:**
1. **SeÅ™adÃ­me** intervaly podle poÄÃ¡teÄnÃ­ho bodu
2. **Inicializujeme** vÃ½sledek s prvnÃ­m intervalem
3. **Projdeme** zbÃ½vajÃ­cÃ­ intervaly a buÄ je slouÄÃ­me, nebo pÅ™idÃ¡me novÃ½
4. **SlouÄenÃ­:** pokud konec poslednÃ­ho â‰¥ zaÄÃ¡tek aktuÃ¡lnÃ­ho

### ğŸ¯ **KlÃ­ÄovÃ© pozorovÃ¡nÃ­:**
- Po seÅ™azenÃ­ staÄÃ­ porovnÃ¡vat jen s poslednÃ­m intervalem ve vÃ½sledku
- Dva intervaly `[a,b]` a `[c,d]` se pÅ™ekrÃ½vajÃ­, pokud `b â‰¥ c` (po seÅ™azenÃ­)
- SlouÄenÃ½ interval: `[min(a,c), max(b,d)]`

```python
class Solution3(object):
    def merge(self, intervals):
        """
        HlavnÃ­ Å™eÅ¡enÃ­ - seÅ™azenÃ­ + greedy merge
        ÄŒasovÃ¡ sloÅ¾itost: O(n log n), ProstorovÃ¡: O(1) aÅ¾ O(n)
        """
        if not intervals:
            return []
        
        # SeÅ™adÃ­me podle poÄÃ¡teÄnÃ­ho bodu
        intervals.sort(key=lambda x: x[0])
        
        # Inicializujeme vÃ½sledek s prvnÃ­m intervalem
        merged = [intervals[0]]
        
        # Projdeme zbÃ½vajÃ­cÃ­ intervaly
        for current in intervals[1:]:
            last = merged[-1]
            
            # Pokud se aktuÃ¡lnÃ­ interval pÅ™ekrÃ½vÃ¡ s poslednÃ­m
            if current[0] <= last[1]:
                # SlouÄÃ­me je - rozÅ¡Ã­Å™Ã­me konec poslednÃ­ho intervalu
                last[1] = max(last[1], current[1])
            else:
                # NepÅ™ekrÃ½vajÃ­ se - pÅ™idÃ¡me novÃ½ interval
                merged.append(current)
                
        return merged

# ğŸ”„ AlternativnÃ­ Å™eÅ¡enÃ­ s explicitnÃ­m slouÄenÃ­m
class Solution3_Alternative(object):
    def merge(self, intervals):
        """
        AlternativnÃ­ pÅ™Ã­stup s explicitnÃ­m vytvÃ¡Å™enÃ­m novÃ½ch intervalÅ¯
        """
        if not intervals:
            return []
            
        intervals.sort()  # SeÅ™azenÃ­ podle prvnÃ­ho prvku (default)
        result = []
        
        for interval in intervals:
            # Pokud je result prÃ¡zdnÃ½ nebo se interval nepÅ™ekrÃ½vÃ¡
            if not result or result[-1][1] < interval[0]:
                result.append(interval)
            else:
                # PÅ™ekrÃ½vÃ¡ se - slouÄÃ­me
                result[-1] = [result[-1][0], max(result[-1][1], interval[1])]
                
        return result

# ğŸ¯ Å˜eÅ¡enÃ­ s detailnÃ­m vysvÄ›tlenÃ­m krokÅ¯
class Solution3_Verbose(object):
    def merge(self, intervals):
        """
        Verbose verze pro pochopenÃ­ algoritmu
        """
        print(f"ğŸ“¥ Vstup: {intervals}")
        
        if not intervals:
            return []
        
        # Krok 1: SeÅ™azenÃ­
        intervals.sort(key=lambda x: x[0])
        print(f"ğŸ“Š SeÅ™azeno: {intervals}")
        
        merged = [intervals[0]]
        print(f"ğŸ¯ Inicializace: {merged}")
        
        # Krok 2: ProchÃ¡zenÃ­ a sluÄovÃ¡nÃ­
        for i, current in enumerate(intervals[1:], 1):
            last = merged[-1]
            print(f"\nğŸ” Krok {i}: ZpracovÃ¡vÃ¡m {current}")
            print(f"    PoslednÃ­ v merged: {last}")
            
            if current[0] <= last[1]:
                # PÅ™ekrÃ½vajÃ­ se
                old_end = last[1]
                last[1] = max(last[1], current[1])
                print(f"    âœ… PÅ™ekrÃ½vajÃ­ se! SlouÄeno: {last}")
                print(f"    ğŸ“ˆ Konec zmÄ›nÄ›n z {old_end} na {last[1]}")
            else:
                # NepÅ™ekrÃ½vajÃ­ se
                merged.append(current)
                print(f"    âŒ NepÅ™ekrÃ½vajÃ­ se. PÅ™idÃ¡n novÃ½: {current}")
            
            print(f"    ğŸ“‹ AktuÃ¡lnÃ­ stav: {merged}")
        
        print(f"\nğŸ‰ FinÃ¡lnÃ­ vÃ½sledek: {merged}")
        return merged

# ğŸ§ª KomplexnÃ­ testovÃ¡nÃ­
def test_merge_intervals():
    print("ğŸ”„ TESTOVÃNÃ BONUS ÃšLOHY: Merge Intervals")
    print("=" * 60)
    
    sol1 = Solution3()
    sol_alt = Solution3_Alternative()
    
    test_cases = [
        ([[1,3], [2,6], [8,10], [15,18]], [[1,6], [8,10], [15,18]]),
        ([[1,4], [4,5]], [[1,5]]),
        ([[4,7], [1,4]], [[1,7]]),
        ([[1,4], [0,4]], [[0,4]]),
        ([[1,4], [2,3]], [[1,4]]),
        ([[1,4], [0,0]], [[0,0], [1,4]]),
        ([], []),
        ([[1,1]], [[1,1]])
    ]
    
    for i, (intervals, expected) in enumerate(test_cases, 1):
        # Testujeme na kopiÃ­ch, protoÅ¾e sort mÄ›nÃ­ pÅ¯vodnÃ­ pole
        result1 = sol1.merge([x[:] for x in intervals])
        result_alt = sol_alt.merge([x[:] for x in intervals])
        
        print(f"Test {i}: {intervals}")
        print(f"  OÄekÃ¡vÃ¡no: {expected}")
        print(f"  HlavnÃ­ Å™eÅ¡enÃ­: {result1} {'âœ…' if result1 == expected else 'âŒ'}")
        print(f"  AlternativnÃ­: {result_alt} {'âœ…' if result_alt == expected else 'âŒ'}")
        print()
    
    # Demonstrace verbose verze na jednom pÅ™Ã­kladu
    print("ğŸ“š DETAILNÃ PRÅ®CHOD algoritmu:")
    print("-" * 40)
    sol_verbose = Solution3_Verbose()
    sol_verbose.merge([[1,3], [2,6], [8,10], [15,18]])
    
    print("\nğŸ“Š AnalÃ½za sloÅ¾itosti:")
    print("  ÄŒasovÃ¡ sloÅ¾itost: O(n log n) - dominuje Å™azenÃ­")
    print("  ProstorovÃ¡ sloÅ¾itost: O(1) aÅ¾ O(n) podle implementace Å™azenÃ­")
    print("  NejhorÅ¡Ã­ pÅ™Ã­pad: vÅ¡echny intervaly se pÅ™ekrÃ½vajÃ­ â†’ jeden vÃ½slednÃ½ interval")
    print("  NejlepÅ¡Ã­ pÅ™Ã­pad: Å¾Ã¡dnÃ© pÅ™ekryvy â†’ pÅ¯vodnÃ­ pole (seÅ™azenÃ©)")

# SpuÅ¡tÄ›nÃ­ testÅ¯
test_merge_intervals()
```
