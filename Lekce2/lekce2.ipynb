### 📚 **Římské číslice:**
| Symbol | Hodnota |
|--------|---------|
| I      | 1       |
| V      | 5       |
| X      | 10      |
| L      | 50      |
| C      | 100     |
| D      | 500     |
| M      | 1000    |

### 🎯 **Klíčová pravidla:**
- **Normálně:** číslice se píší od největší k nejmenší (např. `XII = 12`)
- **Výjimky** (odčítání):
  - `I` před `V` (5) nebo `X` (10) → `IV = 4`, `IX = 9`
  - `X` před `L` (50) nebo `C` (100) → `XL = 40`, `XC = 90`
  - `C` před `D` (500) nebo `M` (1000) → `CD = 400`, `CM = 900`

### 📊 **Příklady:**

**Příklad 1:**
```

Vysvětlení: III = I + I + I = 1 + 1 + 1 = 3
```

**Příklad 2:**
```
Vstup: s = "LVIII"
Výstup: 58
Vysvětlení: L + V + III = 50 + 5 + 3 = 58
```

**Příklad 3:**
```
Vstup: s = "MCMXCIV"
Výstup: 1994
Vysvětlení: M + CM + XC + IV = 1000 + 900 + 90 + 4 = 1994
```

4. Jinak přičti hodnotu symbolu

## 🔄 BONUS: Sloučení překrývajících se intervalů

**Obtížnost:** 🟡 Medium | **Časová složitost:** O(n log n) | **Prostorová složitost:** O(1)

### 📋 **Zadání:**
Dostaneš pole intervalů, kde `intervals[i] = [start_i, end_i]`. **Slouč všechny překrývající se intervaly** a vrať pole nepřekrývajících se intervalů, které pokrývají všechny vstupní intervaly.

### 🎯 **Kdy se intervaly překrývají?**
- Intervaly `[a, b]` a `[c, d]` se překrývají, pokud `b >= c` (po seřazení podle začátku)

### 📊 **Příklady:**

**Příklad 1:**
```
Vstup: intervals = [[1,3], [2,6], [8,10], [15,18]]
Výstup: [[1,6], [8,10], [15,18]]
Vysvětlení: Intervaly [1,3] a [2,6] se překrývají → sloučíme na [1,6]
```

**Příklad 2:**
```
Vstup: intervals = [[1,4], [4,5]]
Výstup: [[1,5]]
Vysvětlení: Intervaly [1,4] a [4,5] se "dotýkají" → sloučíme na [1,5]
```

**Příklad 3:**
```
Vstup: intervals = [[4,7], [1,4]]
Výstup: [[1,7]]
Vysvětlení: Po seřazení [1,4], [4,7] → sloučíme na [1,7]
```

1. **Seřaď** intervaly podle počátečního bodu
